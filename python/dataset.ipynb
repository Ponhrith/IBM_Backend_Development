```python
# Ensure you have installed pandas and openpyxl
# pip install pandas openpyxl

import pandas as pd

# Reading from a local CSV file
csv_path = "./TopSellingAlbums.csv"  # Replace with your local CSV file path
df_csv = pd.read_csv(csv_path)

print("First 5 rows of the CSV data:")
print(df_csv.head())

# Reading from a local Excel file
excel_path = "./TopSellingAlbums.xlsx"  # Replace with your local Excel file path
df_excel = pd.read_excel(excel_path)

print("\nFirst 5 rows of the Excel data:")
print(df_excel.head())

# Accessing specific columns and rows
print("\nAccessing the 'Length' column:")
print(df_csv[['Length']])

print("\nAccessing multiple columns ('Artist', 'Length', 'Genre'):")
print(df_csv[['Artist', 'Length', 'Genre']])

print("\nAccessing a specific cell (1st row, 1st column):")
print(df_csv.iloc[0, 0])

# Using slicing to access a range of rows and columns
print("\nSlicing rows and columns:")
print(df_csv.iloc[0:2, 0:3])
```
